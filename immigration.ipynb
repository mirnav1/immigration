```python
#Compare the trend of top 5 countries that contributed the most to immigration to Canada.

#save changes to original df _can
inplace = True

#get top 5 countries
df_can.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_top5 = df_can.head(5)

#transpose the dataframe
df_top5 = df_top5[years].transpose()

print(df_top5)
```

```python
#area plot
df_can.sort_values(['Total'],ascending=False,axis=0,inplace=True)

df_top5 = df_can.head()
df_top5 = df_top5[years].transpose()

df_top5.head()
```

```python
#countries that contributed the least to immigration to Canada from 1980 to 2013
df_least5 = df_can.tail(5)

df_least5 = df_least5[years].transpose()
df_least5.head()

df_least5.index = df_least5.index.map(int)
df_least5.plot(kind='area',alpha=0.55, stacked=False, figsize=(20,10))

plt.title('Immigration trend of 5 countries with least contribution to immigration')
plt.ylabel('Number of immigrants')
plt.xlabel('Years')

plt.show()
```

```python
#What is the immigration distribution for Denmark, Norway, and Sweden for years 1980 - 2013?
df_can1 = df_can.loc[['Denmark','Norway','Sweden'],years].transpose()
df_can1.head()
```

```python
df_can1.plot(kind='hist',figsize=(10,6))

plt.title('Histogram of immigration from Denmark,Norway and Sweden from 1980 - 2013')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')

plt.show()
```

```python
#increase the bin size to 15 by passing in bins parameter;
#set transparency to 60% by passing in alpha parameter;
#label the x-axis by passing in x-label parameter;
#change the colors of the plots by passing in color parameter.

#x-tick values
count,bin_edges = np.histogram(df_can1,15)

#un-stacked histogram
df_can1.plot(kind='hist',
        figsize=(10,6),
        bins=15,
        alpha=0.6,
        xticks=bin_edges,
        color=['coral','darkslateblue','mediumseagreen']
        )
plt.title('Histogram of immigration from Denmark,Norway and Sweden from 1980-2013')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')

plt.show()
```

```python
#stacked histogram
count,bin_edges=np.histogram(df_can1,15)
xmin = bin_edges[0]-10
xmax=bin_edges[-1]+10

df_can1.plot(kind='hist',
            figsize=(10,6),
            bins=15,
            xticks=bin_edges,
            color=['coral','darkslateblue','mediumseagreen'],
            stacked=True,
            xlim=(xmin,xmax)
            )
plt.title('Histogram of immigration from Denmark,Norway and Sweden from 1980-2013')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')

plt.show()
```

```python
#The immigration distribution for Greece, Albania, and Bulgaria for years 1980 - 2013

df_cof=df_can.loc[['Greece','Albania','Bulgaria'],years] #Countries Of Interest
df_cof = df_cof.transpose()

count,bin_edges = np.histogram(df_cof,15)

df_cof.plot(kind='hist',
            figsize=(10,6),
            bins=15,
            alpha=0.35,
            xticks=bin_edges,
            color=['coral','darkslateblue','mediumseagreen']
            )
plt.title('Histogram of immigration from Greece, Albania and Bulgaria from 1980-2013')
plt.ylabel('Number of years')
plt.xlabel('Number of immigrants')

plt.show()
```

```python
#horizontal bar plot
df_can.sort_values(by='Total',ascending=True,inplace=True)

df_top15 = df_can['Total'].tail(15)
df_top15
```

```python
df_top15.plot(kind='barh',figsize=(12,12),color='steelblue')
plt.xlabel('Number of immigrants')
plt.title('Top 15 countries contributing to the immigration to Canada between 1980-2013')

for index,value in enumerate(df_top15):
    label=format(int(value),',')
    plt.annotate(label,xy=(value-47000,index-0.10),color='white')

plt.show()
```
